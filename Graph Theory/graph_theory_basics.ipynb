```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g = nx.Graph()

```

```python
g.add_node(1)
g.add_node(2)
g.add_node(3)
g.add_node(4)
g.add_node(5)
```

```python
g.node()
```

```python
g.nodes()
```

```python
g.add_edge(1,2)
g.add_edge(1,3)
g.add_edge(4,6)
g.add_edge(5,4)
g.add_edge(2,3)
g.add_edge(2,6)
```

```python
g.edges()
```

```python
nx.draw(g,with_labels=1)
```

```python
nx.adjacency_data(g)
```

```python
nx.draw_circular(g,with_labels=1)
```
