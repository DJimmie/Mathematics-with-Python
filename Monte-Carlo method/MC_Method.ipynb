```python
from scipy import linalg
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import barcode
```

```python
pi_value=[]
n_sample=[]
num_samp=100
for n in range(1,num_samp,1):
    x=np.random.rand(n,2)

    inside=x[np.sqrt(x[:,0]**2+x[:,1]**2)<1]
    estimate=4*len(inside)/len(x)
#     print(f'Estimate of pi: {estimate}')
    pi_value.append(estimate)
    n_sample.append(n)

fig, axes = plt.subplots(1, 2, figsize=(10,3))
axes[0].scatter(x[:,0],x[:,1],s=.5, c='red')
axes[0].scatter(inside[:,0],inside[:,1],s=.5, c='blue')

axes[1].plot(n_sample, pi_value)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)

print(f'Estimate of pi: {estimate}')
print(f'Number of iterations: {num_samp}')
plt.show()






```
